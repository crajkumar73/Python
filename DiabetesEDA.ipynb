```python
df2 = df[['race',
 'gender',
 'age',
 'weight',
 'time_in_hospital',
 'medical_specialty',
 'num_lab_procedures',
 'num_procedures',
 'number_emergency',
 'number_inpatient']]
sns.pairplot(df2)
```

```python
df_num = df.select_dtypes(include = ['float64', 'int64'])
df_num.head()
df_num.hist(figsize=(16, 20), bins=50, xlabelsize=8, ylabelsize=8)
```

```python
df_not_num = df.select_dtypes(include = ['object'])
df_not_num.head()
fig, axes = plt.subplots(round(len(df_not_num.columns) / 2), 2, figsize=(10, 80))

for i, ax in enumerate(fig.axes):
    if i < len(df_not_num.columns):
        ax.set_xticklabels(ax.xaxis.get_majorticklabels(), rotation=45)
        sns.countplot(x=df_not_num.columns[i], alpha=0.9, data=df_not_num, ax=ax)

#fig.tight_layout()
```
